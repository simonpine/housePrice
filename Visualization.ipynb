```python
df_filtered = df[(df['SalePrice'] <= 450000) & (df['GrLivArea'] <= 4000)]
#Tabla sin los datos que pueden llegar a interferir
df_filtered.describe()
```

```python
sns.scatterplot(data=df_filtered, x="SalePrice", y="GrLivArea", size='YearBuilt')
sns.rugplot(data=df_filtered, x="SalePrice", y="GrLivArea")
```

```python
lis = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']


for a in lis:
    df_filtered[a] = (df_filtered[a] - df_filtered[a].min()) / (df_filtered[a].max() - df_filtered[a].min())
# df_filtered['TrainPrice'] = (df_filtered['SalePrice'] - df_filtered['SalePrice'].min()) / (df_filtered['SalePrice'].max() - df_filtered['SalePrice'].min())
# df_filtered['TrainGrLivArea'] =  (df_filtered['GrLivArea'] - df_filtered['GrLivArea'].min()) / (df_filtered['GrLivArea'].max() - df_filtered['GrLivArea'].min())
# df_filtered['TrainYearBuilt'] =  (df_filtered['YearBuilt'] - df_filtered['YearBuilt'].min()) / (df_filtered['YearBuilt'].max() - df_filtered['YearBuilt'].min())

# df_filtered['TrainYearBuilt']
```

```python

K = range(1, 15)

distortions = []
inertias = []

for k in K:
    kmeanModel = KMeans(n_clusters=k).fit(df_filtered[['SalePrice' , 'GrLivArea', 'YearBuilt']])
    kmeanModel.fit(df_filtered[['SalePrice' , 'GrLivArea', 'YearBuilt']])
 
    inertias.append(kmeanModel.inertia_)

sns.lineplot(data=K, x=K, y=inertias)
sns.scatterplot(data=K, x=K, y=inertias)
```

La grafica muestra el impacto que tiene el numero de clusters sobre como se divide el dataset, con ello se puede decir que 4 centroides llegan a ser suficientemente relevantes, sin caer en el desuso

```python
kmeans = KMeans(n_clusters=4)
df_filtered['Groups'] = kmeans.fit_predict(df_filtered[['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt']])
```

```python
sns.scatterplot(data=df_filtered, x="GrLivArea", y="SalePrice", hue='Groups', size='YearBuilt')
sns.rugplot(data=df_filtered, x="GrLivArea", y="SalePrice", hue='Groups')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=100, alpha=0.5);
```

En la grafica se puede apreciar como se generan 4 principales grupos teniendo en cuenta el area de vivienda, el precio y el ano de construccion. No se ve perfectamente dividido, esto sucede porque se esta intentando mostrar en 2 dimenciones 3 variables distintas, siendo el tamano de los puntos la variable 'YearBuilt'. Con esta divicion por grupos podemos sacar algunas ideas, por ejemplo existe un grupo, el 3, el cual se caracteriza por tener un espacio aplio de vivienda, y haber sido construido hace menos anos, y por ende tener un precio superior a la media. Por otro lado se encuentran los grupos 2 y 1, los cuales llaman la atencion por tener un tamano y precio muy similar y alrededor de la media, pero se diferencian por se mas nuevas casas del grupo 2.

```python
cols = ['SalePrice', 'OverallQual', 'GrLivArea', 'GarageCars', 'TotalBsmtSF', 'FullBath', 'YearBuilt', 'Groups']
sns.pairplot(df_filtered[cols], hue='Groups')
```

En la tabla anterior se aprecia como la relacion entre el area de la casa y el precio tienen una gran relacion directa, lo que no pasa con el area y el ano de construccion. Pero el ano de construccion si aporta una buena relacion con el precio, por lo que tanto el ano como el area nos permitirian determinar el precio de una vivienda.

```python
X_train, X_test, y_train, y_test = train_test_split(df_filtered[['GrLivArea']], 
                                                    df_filtered['SalePrice'], test_size=0.2)
```

```python

linear_model= LinearRegression()

linear_model.fit(X_train, y_train)

predictions = linear_model.predict(X_test)

print("Coeficiente:", linear_model.coef_)
# The mean squared error
print("Coeficiente de error promedio: %.2f" % mean_squared_error(y_test, predictions))
# The coefficient of determination: 1 is perfect prediction
print("Coeficiente de acierto: %.2f" % r2_score(y_test, predictions))


sns.scatterplot(data=df_filtered, x="GrLivArea", y="SalePrice", hue='Groups')
plt.plot(X_test, predictions,color='k') 
```
